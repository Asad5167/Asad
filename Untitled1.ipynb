```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import curve_fit
```

```python
# Load groundwater data CSV file for Harris County
data = pd.read_csv('C:\Users\asadh\Downloads\MiniP2')
```

```python
# Preview the data
print(data.head())
```

```python
# Filter data for necessary columns (e.g., Well ID, Latitude, Longitude, Hydraulic Head)
columns_needed = ['Well ID', 'Latitude', 'Longitude', 'Hydraulic Head']
filtered_data = data[columns_needed].dropna()
```

```python
# Visualize the locations on a 2D map
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['Longitude'], filtered_data['Latitude'], c=filtered_data['Hydraulic Head'], cmap='viridis')
plt.colorbar(label='Hydraulic Head (m)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Hydraulic Head Distribution in Harris County')
plt.show()
```

```python
# Fit a 2D plane to the hydraulic heads to compute gradient
def plane_equation(coords, a, b, c):
    x, y = coords
    return a * x + b * y + c

x = filtered_data['Longitude'].values
y = filtered_data['Latitude'].values
z = filtered_data['Hydraulic Head'].values
```

```python
# Fit the plane
params, _ = curve_fit(plane_equation, (x, y), z)
```

```python
# Extract gradient coefficients
a, b, c = params
gradient_vector = np.array([a, b])\
```

```python
print(f'Gradient coefficients: a = {a}, b = {b}')
```

```python
# 3D Visualization
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', label='Data Points')
```

```python
# Plot fitted plane
X, Y = np.meshgrid(np.linspace(min(x), max(x), 100), np.linspace(min(y), max(y), 100))
Z = plane_equation((X, Y), *params)
ax.plot_surface(X, Y, Z, color='yellow', alpha=0.5)
```

```python
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Hydraulic Head')
plt.title('3D Plot of Fitted Hydraulic Head Plane')
plt.show()
```

```python
# Calculate and print gradient magnitude and direction
gradient_magnitude = np.linalg.norm(gradient_vector)
gradient_direction = np.arctan2(b, a)
```

```python
print(f'Gradient magnitude: {gradient_magnitude}')
print(f'Gradient direction (radians): {gradient_direction}')
```
